```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('thefan.csv', dtype=object)
```

```python
df = df.dropna(how='all', axis=1)
```

```python
string = '20160629'
```

```python
pd.to_datetime(string, format='%Y%m%d')
```

```python
def convert_to_datetime(string):
    match = re.match('[^-]*', string).group(0)
    return pd.to_datetime(match, infer_datetime_format=True, errors='coerce')
```

```python
df['DATE'] = df['INCIDENT_NUMBER'].apply(lambda x: convert_to_datetime(x))
```

```python
df['DATE'].describe()
```

```python
cols = list(df)

cols.insert(0, cols.pop(cols.index('DATE')))

df = df[cols]
```

```python
df = df.drop('INCIDENT_NUMBER', 1) # These values are not unique, so this column is useless
```

```python
# Order the dataset according to date
df = df.sort_values(by=['DATE'])
```

```python
# Add ID column to keep track of incidents
df.reset_index()

df['ID'] = df.index

cols = list(df)
cols.insert(0, cols.pop(cols.index('ID')))
df = df[cols]
```

```python
df
```

```python
# weirdly necessary to delete all rows with dates > 2020
cap = pd.to_datetime('2020')
df = df[df['DATE'] < cap]
```

```python
df.to_csv('the_fan_output.csv', index=False)
```
